```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
a=pd.read_csv('1.csv', delimiter=',', sep='\t')
a
```

## droping 1 column which is un necessary

```python
a.drop(a.columns[0],axis=1, inplace=True)
a
```

```python
a.head()
```

## Cleaning car names of space and characters

```python
a=a.replace('"','',regex=True).replace('\t',"",regex=True)
# replace('\"',"")
a
```

## Checking null values

```python
a.info()
```

## Droping null rows

```python
a.dropna(axis=1,inplace=True)
a
```

```python
a.isnull().sum()
```

## Null values as '?'

```python
print(a[a['horsepower:']== '?'])
```

## Null value '?' handled

```python
b=a[a['horsepower:']!= '?']
b
```

```python
a['car_name:'].value_counts()
```

## Grouping cars together and its attributes

```python
c=a.groupby('car_name:')
print(c.get_group('ford country squire (sw)'))
amc_matador=c.get_group('amc matador')
ford_pinto=c.get_group('ford pinto')
toyota_corolla =c.get_group('toyota corolla')
chevrolet_impala=c.get_group('chevrolet impala')
amc_matador.describe()
li=['mpg:','horsepower:','cylinders:','acceleration:']
a1=[]
a2=[]
a3=[]
a4=[]
for i in li:
    max_value= amc_matador[i].max()
    a1.append(max_value)
for i in li:
    max_value= ford_pinto[i].max()
    a2.append(max_value)
for i in li:
    max_value= toyota_corolla[i].max()
    a3.append(max_value)
for i in li:
    max_value= chevrolet_impala[i].max()
    a4.append(max_value)
```

## Segmenting cars

```python
name=['amc_matador','ford_pinto','toyota_corolla','chevrolet_impala']
a1
mpg=[]
hp=[]
cyc=[]
acc=[]
def u(a,b,c,d,e):
    for i in range(1):
        a.append(b[i])
        c.append(b[i])
        d.append(b[i])
        e.append(b[i])

u(mpg,a1,hp,cyc,acc)
u(mpg,a2,hp,cyc,acc)
u(mpg,a3,hp,cyc,acc)
u(mpg,a4,hp,cyc,acc)
mpg
```

## Performing analysis on Cars to find out best car

```python
import numpy as np
X = np.arange(4)
plt.figure(figsize=(15,8))
plt.bar(X+.00, a1, color = 'b', width = 0.20,label='MPG')
plt.bar(X + 0.20, a2, color = 'g', width = 0.20, label='Horse_power')
plt.bar(X + 0.40, a3, color = 'r', width = 0.20, label='Cylinder')
plt.bar(X + .60, a4, color = 'orange', width = 0.20, label='Acceleration')
plt.xticks(X, name)
plt.legend()
plt.show()
# plt.bar(, data[1], color = 'g', width = 0.25)
```

## By looking at the graph,  Ford Pinto is the best car in terms of MPG, HP, Cylinder & Acceleration. Amc_Matador came in second.

```python
b['mpg:'].plot.kde(xlim=(b['mpg:'].min(),b['mpg:'].max()),title='Distribution for MPG')
plt.axvline(b['mpg:'].mean() ,color="black", label="mean")
plt.axvline(b['mpg:'].mode()[0],color="red", label="mode")
plt.axvline(b['mpg:'].median(),color="green",label="median")
plt.legend()
plt.show()
```
